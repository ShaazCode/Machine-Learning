Example Scenario

Imagine you have two types of fruits:

Class A: Apples

Class B: Oranges

Suppose you are using two features:

Weight

Color score (0 means fully green, 10 means fully orange)

You plot the data like points on a graph.

     |                            Orange (B)
     |                       
Color|                 
Score|      Apple (A)               Hyperplane (decision boundary)
     |                      ------------------------------
     |        
     |
     +------------------------------------------------------
                      Weight


What SVM does here

It tries to find a straight line (hyperplane) that separates Apples from Oranges.

It also draws two parallel margin lines on both sides of this separating line.

The closest points from both classes that touch these margin lines are called support vectors.
